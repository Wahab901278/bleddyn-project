```python
fig,axes=plt.subplots(3,3,figsize=(18,14))
axes=axes.ravel()

for index,column in enumerate(numeric_columns.columns):
    if index < len(axes):
        axes[index].hist(df[column],bins=50)
        axes[index].set_title(f'Distribution of {column}')
```

```python
categorical_columns=['product_name',
       'main_category', 'sub_category', 'brand', 'payment_method', 'region',
       'customer_age_group', 'device_type']

fig,axes=plt.subplots(3,3,figsize=(18,14))
axes=axes.ravel()

for index,column in enumerate(categorical_columns):
    if index < len(axes):
        value_counts=df[column].value_counts()
        if len(value_counts.index)<10:
            axes[index].bar(value_counts.index,value_counts.values)
            axes[index].set_title(f"Distribution of {column}")
```
